# Chapter 3: Hello, little turtles!

<h1> First Turtle </h1>

```python
import turtle
window = turtle.Screen() # Creates a playground for turtles
# We change window properties
colorVentana = input("What color do you want? (peach puff): ")
window.bgcolor(colorVentana)
window.title("Hola, Peque")

chui = turtle.Turtle() # Create a turtle

chui.speed(1)  # Speed between 0 and 10
colorChui = input("What color turtle do you want?: ")
chui.color(colorChui)
chuiWidth = input("What size do you want for Chui?: ")
chui.pensize(int(chuiWidth))
chui.forward(50) # Move forward by 50 units
chui.left(90) # Turn by 90 degrees
chui.forward(30) # Complete the second side of a rectangle


squirtle = turtle.Turtle() # Let's create another one
squirtle.speed(1)
squirtle.color("blue")
squirtle.left(180)
squirtle.forward(50)

window.mainloop() # Wait for user to close window
```

<h1> First Turtle with For </h1>

```python
import turtle
window = turtle.Screen() # Creates a playground for turtles
# We change window properties
window.bgcolor("peach puff")
window.title("Hello")

t1 = turtle.Turtle() # Create a turtle
t1.speed(1)  # Speed between 0 and 10

clrs = ["yellow", "red", "purple", "blue"]

for i in clrs:
  t1.color(i)
  t1.forward(50) # Move forward by 50 units
  t1.left(90) # Turn by 90 degrees

window.mainloop() # Wait for user to close window
```

<h1> Extra Methods and Tricks </h1>

* t1.penup() -> Moves but does not draw anything
* t1.pendown() -> Allows drawing again
* t1.shape(“”) -> Tells the turtle its shape (arrow, blank, circle,
classic, square, triangle, turtle)
* t1.stamp() -> Leaves marks where it goes, the mark has the shape of the turtle
* t1.hideturtle() -> To hide the turtle, but continue drawing
* t1.showturtle() -> To show it again

2. A drunk pirate makes a random turn and then takes 100 steps forward, makes another random
turn, takes another 100 steps, turns another random amount, etc. A social science student
records the angle of each turn before the next 100 steps are taken. Her experimental data is
[160, -43, 270, -97, -43, 200, -940, 17, -86]. (Positive angles are counter-clockwise.)
Use a turtle to draw the path taken by our drunk friend.
3. Enhance your program above to also tell us what the drunk pirate’s heading is after he has
finished stumbling around. (Assume he begins at heading 0).

```python
import turtle
window = turtle.Screen()
window.bgcolor("peach puff")
window.title("Hello")

t1 = turtle.Turtle()
t1.speed(10)

step = [160, -43, 270, -97, -43, 200, -940, 17, -86]

for i in step:
  t1.forward(100)
  t1.left(i)

window.mainloop()
```
