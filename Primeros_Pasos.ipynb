## Data Science

You can harness the full power of Python's most popular libraries to analyze and visualize data. The code cell below uses <strong>NumPy</strong> to generate random data and <strong>Matplotlib</strong> to visualize it. To edit the code, just click on the cell.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```
