<h1> A tuple cannot be modified </h1>

To do this, we must split it and add the new one where appropriate.

```python
julia = julia[:3] + ("Eat Pray Love", 2010) + julia[5:]
```

<h1> We can separate the tuples </h1>

```python
b = ("Bob", 19, "CS")
(name, age, studies) = b
name
```

<h1> We can change values with tuples </h1>

```python
a = 1
b = 2
(a, b) = (b, a)
print(a)
print(b)
```
