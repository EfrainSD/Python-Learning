<h1> Split a String </h1>


```python
s = "Pirates of the Caribbean"
print(s[0:7])

print(s[11:14])

print(s[15:24])

friends = ["Joe", "Zoe", "Brad", "Angelina", "Zuki", "Thandi", "Paris"]
print(friends[2:4])
```

Another example of division, but in a different way

```python
fruit = "banana"
print(fruit[:3])
print(fruit[3:])
print(fruit[3:999])
```

<h1>The best way to write strings</h1>

This is quite important, especially when you want to print values that may change.

```python
s1 = "His name is {0}!".format("Arthur")
print(s1)

name = "Alice"
age = 10
s2 = "I am {1} and I am {0} years old.".format(age, name)
print(s2)

n1 = 4
n2 = 5
s3 = "2**10 = {0} and {1} * {2} = {3:f}".format(2**10, n1, n2, n1 * n2)
print(s3)
```
